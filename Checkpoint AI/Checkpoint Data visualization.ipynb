```python
CabinC = data_tita["Cabin"].value_counts().keys()[0]
data_tita["Cabin"].fillna(CabinC, inplace = True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data_tita["Cabin"])
data_tita["CabinEncoder"] = le.transform(data_tita["Cabin"])
```
